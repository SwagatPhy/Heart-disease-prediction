```python
from sklearn.ensemble import VotingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import RandomForestClassifier

voting = VotingClassifier(
    estimators=[
        ('xgb', XGBClassifier(random_state=42)),
        ('lgbm', LGBMClassifier(random_state=42)),
        ('rf', RandomForestClassifier(random_state=42))
    ],
    voting='soft'  # uses probabilities
)
voting.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Basic Metrics
y_val_pred = voting.predict(x_test)
y_val_proba = voting.predict_proba(x_test)[:,1]

print("Val Accuracy:", accuracy_score(y_test, y_val_pred))
print("Val ROC AUC:", roc_auc_score(y_test, y_val_proba))

# 2. Detailed Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_val_pred))

# 3. Confusion Matrix
cm = confusion_matrix(y_test, y_val_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Disease', 'Disease'],
            yticklabels=['No Disease', 'Disease'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# 4. Compare all models
models = {
    'XGBoost': xgb,
    'Voting Classifier': voting,
}

print("\n--- Model Comparison ---")
for name, model in models.items():
    y_pred = model.predict(x_test)
    y_proba = model.predict_proba(x_test)[:,1]
    print(f"{name}:")
    print(f"  Accuracy : {accuracy_score(y_test, y_pred):.4f}")
    print(f"  ROC AUC  : {roc_auc_score(y_test, y_proba):.4f}")
    print()
```

```python
# build cnn model
import tensorflow as tf

model = tf.keras.Sequential([
  # tf.keras.Input(shape=(x_train.shape[0]))
  tf.keras.layers.Dense(units=64, activation='relu', input_shape=(x_train.shape[1],)),
  tf.keras.layers.Dense(units=32, activation='relu'),
  tf.keras.layers.Dense(units=16, activation='relu'),
  tf.keras.layers.Dense(units=1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model_history = model.fit(x_train, y_train, epochs=10, batch_size=64)
model_history
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score

# Convert probabilities to binary
y_train_pred = (model.predict(x_train) > 0.5).astype(int)
y_val_pred = (model.predict(x_test) > 0.5).astype(int)

# For ROC AUC keep probabilities
y_val_proba = model.predict(x_test)

print("Train Accuracy:", accuracy_score(y_train, y_train_pred))
print("Val Accuracy:", accuracy_score(y_test, y_val_pred))
print("Val ROC AUC:", roc_auc_score(y_test, y_val_proba))
```
